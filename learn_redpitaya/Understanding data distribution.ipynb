# Understanding Person data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm

# generate some example data
data = df.iloc[1, 1:]

data = data.to_numpy()



# create a histogram
plt.hist(data, bins=10)
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data Distribution')
plt.show()

# create a density plot
sns.kdeplot(data)
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.title('Density Plot of Data Distribution')
plt.show()


# create a boxplot
sns.boxplot(data)
plt.xlabel('Data Values')
plt.title('Boxplot of Data Distribution')
plt.show()

# create a violin plot
sns.violinplot(data)
plt.xlabel('Data Values')
plt.title('Violin Plot of Data Distribution')
plt.show()

# create a CDF
counts, bin_edges = np.histogram(data, bins=10, density=True)
cdf = np.cumsum(counts)
plt.plot(bin_edges[1:], cdf)
plt.xlabel('Data Values')
plt.ylabel('Cumulative Probability')
plt.title('CDF of Data Distribution')
plt.show()

# create a swarmplot
sns.swarmplot(data)
plt.xlabel('Data Values')
plt.title('Swarmplot of Data Distribution')
plt.show()


# # create a heatmap
data_reshaped = data.reshape(-1, 1)
plt.imshow(data_reshaped, cmap='YlGnBu', interpolation='nearest')
plt.colorbar()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap of Data Distribution')
plt.show()


# create a PDF
x_axis = np.linspace(data.min(), data.max(), 100)
plt.plot(x_axis, norm.pdf(x_axis, data.mean(), data.std()))
plt.hist(data, density=True, alpha=0.5)
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.title('PDF of Data Distribution')
plt.show()

```

### Saving person and non-person plots as img

```python
import matplotlib.pyplot as plt
import numpy as np



# Set the same limits for the X and Y axes
xmin, xmax = 0, 1000
ymin, ymax = 0, 0.01

x_axis_plots = 2
y_axis_plots = 3

num_of_figures = 50

axis_positions = [[i, j] for i in range(x_axis_plots) for j in range(y_axis_plots)]

colors = ["blue", "green", "cyan", "purple", "magenta", "gold"]



def get_data(indx, x_axis_plots=2, y_axis_plots=3, person_data=True):
    df = pd.read_csv("20k_dataframe_2303.csv", header=None)
    if person_data:
        df_single_class = df[df[0] == 1]
    else:
        df_single_class = df[df[0] == 0]
    return df_single_class.iloc[:, 1:].to_numpy()[indx:indx+(x_axis_plots*y_axis_plots)]




def plot_histrogram(data, index, axs):
    x_indx = axis_positions[index][0]
    y_indx = axis_positions[index][1]
    color = colors[index]
    x_axis_data = np.linspace(data.min(), data.max(), 100)
    axs[x_indx, y_indx].plot(x_axis_data, norm.pdf(x_axis, data.mean(), data.std()), color="red")
    axs[x_indx, y_indx].hist(data, bins=30, density=True, color=color)
    axs[x_indx, y_indx].set_xlim([xmin, xmax])
    axs[x_indx, y_indx].set_ylim([ymin, ymax])



# Add axis labels and a title
# fig.suptitle('Data Distribution (X-axis is the data values & Y-axis is the Probability density)', fontsize=10)



for i in range(num_of_figures):
    fig, axs = plt.subplots(x_axis_plots, y_axis_plots, figsize=(10, 10))
    data = get_data(i, person_data=False)
    for index, data in enumerate(data):
        plot_histrogram(data, index, axs)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    fig.savefig(f"distribution_analysis/20k/non_person/{i}_non_person.svg", format="svg")
    # plt.show()
    plt.close()
```

# Create plot for ultrasonic data received from sensor

```python
import socket
import struct
import pickle
import time
import pandas as pd



buffer_size = 65536

msg_from_client = "-f 1"

bytes_to_send = str.encode(msg_from_client)

server_address_port = ("192.168.128.1", 61231)


# Create a UDP socket at client side
udp_client_socket = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)

# Send to server using created UDP socket

udp_client_socket.sendto(bytes_to_send, server_address_port)

# Set the same limits for the X and Y axes
xmin, xmax = 0, 1000
ymin, ymax = 0, 0.01

x_axis_plots = 2
y_axis_plots = 3

num_of_figures = 50

axis_positions = [[i, j] for i in range(x_axis_plots) for j in range(y_axis_plots)]

colors = ["blue", "green", "cyan", "purple", "magenta", "gold"]




def plot_histrogram(data, index, axs):
    x_indx = axis_positions[index][0]
    y_indx = axis_positions[index][1]
    color = colors[index]
    x_axis_data = np.linspace(data.min(), data.max(), 100)
    axs[x_indx, y_indx].plot(x_axis_data, norm.pdf(x_axis, data.mean(), data.std()), color="red")
    axs[x_indx, y_indx].hist(data, bins=30, density=True, color=color)
    axs[x_indx, y_indx].set_xlim([xmin, xmax])
    axs[x_indx, y_indx].set_ylim([ymin, ymax])

def get_ultrasonic_data():
    packet = udp_client_socket.recv(buffer_size)
    # print(f"Total Received : {len(packet)} Bytes.")
    header_length = int(struct.unpack('@f', packet[:4])[0])
    ultrasonic_data_length = int(struct.unpack('@f', packet[4:8])[0])
    
    ultrasonic_data = []
    for i in struct.iter_unpack('@h', packet[header_length:]):
        ultrasonic_data.append(i[0])
    # print("Ultrasonic Data Length : ", ultrasonic_data_length)
    # print("Ultrasonic Data : ", ultrasonic_data)
    return ultrasonic_data


print("Receiving data and saving the plot...")


for i in range(50):
    indx = 0
    fig, axs = plt.subplots(x_axis_plots, y_axis_plots, figsize=(10, 10))
    while indx <= 5:
        # time.sleep(2)
        data = pd.Series(get_ultrasonic_data())
        plot_histrogram(data, indx, axs)
        indx += 1
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    fig.savefig(f"distribution_analysis/real_time_data/person/{i}_person.svg", format="svg")
    plt.close()

print("Done")
udp_client_socket.close()
```
