```python
import socket
import struct

msg_to_server       = "-a 1"

bytes_to_send         = str.encode(msg_to_server)

rps_address_port   = ("192.168.128.1", 61231)

buffer_size          = 65536

# Create a UDP socket at client side
udp_client_socket = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)
udp_client_socket.sendto(bytes_to_send, rps_address_port)

print("Start")

packet = udp_client_socket.recv(buffer_size)
header_length = int(struct.unpack('@f', packet[:4])[0])
ultrasonic_data_length = int(struct.unpack('@f', packet[4:8])[0])

print(f"Header length {header_length}")
print(f"Ultrasonic Data length {ultrasonic_data_length}")

header_data = []
for i in struct.iter_unpack('@f', packet[:header_length]):
    header_data.append(i[0])
print(len(header_data))

ultrasonic_data = []
for i in struct.iter_unpack('@h', packet[header_length:]):
    ultrasonic_data.append(i[0])
print(len(ultrasonic_data))
```

```python
def get_data_from_sensor(msg_to_server="-a 1"):
    # importing necessary library
    import struct
    import socket
    
    # Defining server and port
    msg_to_server       = msg_to_server
    bytes_to_send         = str.encode(msg_to_server)
    rps_address_port   = ("192.168.128.1", 61231)
    buffer_size          = 65536

    # Create a UDP socket at client side
    udp_client_socket = socket.socket(family=socket.AF_INET, type=socket.SOCK_DGRAM)
    udp_client_socket.sendto(bytes_to_send, rps_address_port)

    print("Start")

    packet = udp_client_socket.recv(buffer_size)
    header_length = int(struct.unpack('@f', packet[:4])[0])
    ultrasonic_data_length = int(struct.unpack('@f', packet[4:8])[0])

    print(f"Header length {header_length}")
    print(f"Ultrasonic Data length {ultrasonic_data_length}")

    header_data = []
    for i in struct.iter_unpack('@f', packet[:header_length]):
        header_data.append(i[0])
    print(len(header_data))

    ultrasonic_data = []
    for i in struct.iter_unpack('@h', packet[header_length:]):
        ultrasonic_data.append(i[0])
    print(len(ultrasonic_data))
    
    return ultrasonic_data

# get_data_from_sensor("-f 1")
get_data_from_sensor()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
import io

num_rows = 10
num_cols = 1
fig, axs = plt.subplots(num_rows, num_cols, sharex=True)
for i in range(num_rows):
    ax = axs[i]
    ax.plot(np.arange(10), np.arange(10)**i)

def on_click(event):

    if not event.inaxes: return
    inx = list(fig.axes).index(event.inaxes)
    buf = io.BytesIO()
    pickle.dump(fig, buf)
    buf.seek(0)
    fig2 = pickle.load(buf) 

    for i, ax in enumerate(fig2.axes):
        if i != inx:
            fig2.delaxes(ax)
        else:
            axes=ax

    axes.change_geometry(1,1,1)
    fig2.show()

fig.canvas.mpl_connect('button_press_event', on_click)

plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig, ax = plt.subplots()

# initialize the plot with empty data
line, = ax.plot([], [])

def init():
    line.set_data([], [])
    return line,

def update(frame):
    x = np.arange(0, 2*np.pi, 0.1)
    y = np.sin(x + frame/10.0)
    line.set_data(x, y)
    return line,

ani = FuncAnimation(fig, update, frames=np.arange(0, 10), init_func=init, blit=True, interval=10)
plt.show()
```
