```python
import threading
import time

start = time.perf_counter()


def do_something():
    print("sleep 1 sec")
    time.sleep(1)
    print("done sleeping")
    
t1 = threading.Thread(target=do_something)
t2 = threading.Thread(target=do_something)


# run the thread
t1.start()
t2.start()

# threads task would be finished before below lines
t1.join()
t2.join()

finish = time.perf_counter()

print(f"Finished {finish -start}")
```

```python
import threading
import time

start = time.perf_counter()


def do_something():
    print("sleep 1 sec")
    time.sleep(1)
    print("done sleeping")

threads = []    
for _ in range(10):
    t = threading.Thread(target=do_something)
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()

finish = time.perf_counter()

print(f"Finished {finish -start}")
```

```python
import threading
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    

threads = []    
for _ in range(10):
    t = threading.Thread(target=do_something, args=[1.5])
    t.start()
    threads.append(t)

for thread in threads:
    thread.join()

finish = time.perf_counter()

print(f"Finished {finish -start}")
```

```python
import concurrent.futures
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    return "done"

with concurrent.futures.ThreadPoolExecutor() as executor:
    f1 = executor.submit(do_something, 1)
    f2 = executor.submit(do_something, 1)
    print(f1.result())
    print(f2.result())

finish = time.perf_counter()

print(f"Finished {finish -start}")
```

```python
import concurrent.futures
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    return "done"

with concurrent.futures.ThreadPoolExecutor() as executor:
    results = [executor.submit(do_something, 1) for _ in range(10)]

for f in concurrent.futures.as_completed(results):
    print(f.result())

finish = time.perf_counter()

print(f"Finished {finish -start}")
```

```python
import concurrent.futures
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    return f"done {arg}"

with concurrent.futures.ThreadPoolExecutor() as executor:
    secs = [5, 4, 3, 2, 1]
    results = executor.map(do_something, secs)
    
    for result in results:
        print(result)
```

## Multiprocessing

```python
import multiprocessing
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    return f"done {arg}"

do_something(1)
```

```python
import multiprocessing
import time

start = time.perf_counter()


def do_something(arg):
    print(f"sleep {arg} sec")
    time.sleep(arg)
    print("done sleeping")
    return f"done {arg}"

p1 = multiprocessing.Process(target=do_something, args=[1])
p2 = multiprocessing.Process(target=do_something, args=[1])

# print(p1)'

p1.start()
p2.start()



finish = time.perf_counter()

print(f"Finished {finish -start}")
```
