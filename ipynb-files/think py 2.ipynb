```python
def factorial(n):
    if n < 1:
        return 1
    else:
        recurse = factorial(n-1)
        result = n * recurse
        return result
```

```python
factorial(6)
```

```python
def fibonacci(n):
    if n <= 1:
        return 1
    a = 1
    b = 1
    print(a)
    for i in range(n):
        temp = b
        b = temp + a
        a = temp
        print(b)

fibonacci(5)
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(6)
```
