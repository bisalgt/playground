```python
import pytest

def func(x):
    return x**2

def test_method(i):
    assert func(i) == 16
    
test_method(3)
```

```python
test_method(4)
```

```python
@pytest.mark.one
def test_method1():
    x = 5
    y = 6
    assert x == y
    
@pytest.mark.two
def test_method2():
    a = 7
    b = 14
    assert a*2 == b
```
