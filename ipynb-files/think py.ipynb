```python
for i in range(1, 21):
    if i == 1 or i % 5 == 0:
        for j in range(1, 21):
            if j == 1 or j % 5 == 0:
                print("+", end=" ")
            else:
                print("-", end=" ")
    else:
        for k in range(1,21):
            if k == 1 or k % 5 == 0:
                print("|", end=" ")
            else:
                print(" ", end=" ")
    print()
```

```python
def polygon(t, length, n):  
    angle = 360/n
    for i in range(n):
        t.fd(length)
        t.lt(angle)
```

```python
def circle(t, radius):
    length = 1
    polygon(t,length, int( 2 * 22*radius/ 7 / length))
```
