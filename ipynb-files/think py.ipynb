```python
def polygon(t, length, n, theta):  
    angle = theta/n
    for i in range(n):
        t.fd(length)
        t.lt(angle)
```

```python
def circle(t, radius):
    length = 1
    polygon(t,length, int( 2 * 22*radius/ 7 / length))
```

```python
import turtle
tut_obj = turtle.Turtle()
```

```python
circle(tut_obj, 60)
```

```python
def arc(t, r, angle):
    length_of_a_side = 1
    total_length_of_arc = 2 * 22 * r * angle / 7 / 360
    number_of_sides = int(total_length_of_arc / length_of_a_side)
    polygon(t, length_of_a_side, number_of_sides, angle)
    
```
