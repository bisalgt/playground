```python
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)

factorial(1)
```

#### using decorator

```python
def printer(a):
    return f"HI i am printer with {a}"

def upper_decorator(func):
    def wrapper(arg):
        value = func(arg).upper()
        return value
    return wrapper


printer("ibshal")      
```

```python
@upper_decorator
def dec_print(a):
    return f"HI i am printer with {a}"

dec_print("ibishal")
```

```python
from copy import copy, deepcopy
list_1 = [1, 2, [3, 5], 4]
## shallow copy
list_2 = copy(list_1) 
list_2[3] = 7
list_2[2].append(6)
print(list_2)    # output => [1, 2, [3, 5, 6], 7]
print(list_1)    # output => [1, 2, [3, 5, 6], 4]
```

```python
list_1 is list_2
```

```python
list_1[2] is list_2[2]
```

```python
## deep copy
list_3 = deepcopy(list_1)
list_3[3] = 8
list_3[2].append(7)
print(list_3)    # output => [1, 2, [3, 5, 6, 7], 8]
print(list_1)
```

```python
list_3 is list_1
```

```python
list_3[2] is list_1[2]
```
