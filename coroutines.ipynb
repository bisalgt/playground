```python
import asyncio
import time

async def say_after(delay, what):
    print('inside funch')
    await asyncio.sleep(delay)
    print(what)

async def main():
    print(f"started at {time.strftime('%X')}")

    await say_after(1, 'hello')
    await say_after(2, 'world')

    print(f"finished at {time.strftime('%X')}")

await main()
```

```python
async def main():
    task1 = asyncio.create_task(
        say_after(1, 'hello'))

    task2 = asyncio.create_task(
        say_after(2, 'world'))

    print(f"started at {time.strftime('%X')}")

    # Wait until both tasks are completed (should take
    # around 2 seconds.)
    await task1
    await task2

    print(f"finished at {time.strftime('%X')}")
await main()
```

```python
import asyncio

async def nested():
    return 42

async def main():
    # Schedule nested() to run soon concurrently
    # with "main()".
    task = asyncio.create_task(nested())

    # "task" can now be used to cancel "nested()", or
    # can simply be awaited to wait until it is complete:
    await task


print(await main())
```

## coroutines- function with special power which allow it to pause and resume its operation

### coroutines can be declared using async

```python
async def main():
    print('hello')
    
main() # when calling coroutines, we don't actually execute a function, but rather it returns a coroutines object
```

### To pause execution of coroutines we use await in coroutines

### To execute a coroutines, we need event loop. Event loop runs in main thread and executes callbacks and tasks in its threads

```python
import asyncio
async def main(): # coroutine
    print('hello coroutines')
    
loop = asyncio.get_event_loop()
# to pass coroutines object we execute coroutine
loop.run_until_complete(main())
loop.close()
```

## to execute coroutines concurrently, 
## we can use ayncio.gather()

```python
import time
import asyncio

       
async def prin_num():
    print('inside prin_num function')
    num = 1
    while num <15:
        print(num)
        num += 1
        await asyncio.sleep(1)

# we have only single thread, can do only one operation at a time due to GIL
async def display_time():
    start_time = time.time()
    print('task one, display time')
#     print(type(start_time))
    dur = 0
    while dur < 15:
        dur = int(time.time() - start_time)
        print('inside while of display time')
        if dur % 3 == 0:
            print(f'{dur} seconds has passed')
        await asyncio.sleep(0.5) # this particular coroutine will be paused and allow other coroutine to run concurrently
         

async def main():
#     task1 = asyncio.create_task(display_time()) # added in python 3.7
    task1 = asyncio.ensure_future(display_time())
    task2 = asyncio.create_task(prin_num())
    print('inside main')
    await asyncio.gather(task2, task1)
    print('end of all tasks')
    
await main()
```

```python
import requests
import time

url = 'https://httpbin.org/uuid'

def fetch(session, url):
    with session.get(url) as response:
        print(response.json()['uuid'])

def main():
    with requests.Session() as session:
        for _ in range(50):
            fetch(session, url)

st = time.perf_counter()
main()
en = time.perf_counter()

print(en-st)
```

```python
import timeit

def timer(number, repeat):
    def wrapper(func):
        runs = timeit.repeat(func, number=number, repeat=repeat)
        print(sum(runs)/len(runs))
    return wrapper
```

```python
import requests
import time
from multiprocessing.pool import Pool

url = 'https://httpbin.org/uuid'

def fetch(session, url):
    with session.get(url) as response:
        pass
#         print(response.json()['uuid'])
@timer(1,1)
def main():
    with Pool() as pool:
        with requests.Session() as session:
#                 fetch(session, url)
            pool.starmap(fetch, [(session, url) for _ in range(50)])
```
