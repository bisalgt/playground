## Regularization

## To remove the problem of over fitting we use regularization, to penalize the weights (theta )

```python
X = pd.concat([pd.Series(np.ones(df.shape[0])), df[['studied', 'slept']]], axis=1)
X['studied2'] = X['studied']**2
X['slept2'] = X['slept']**2
X
```

```python
y = df['pass'].copy()
y
```

```python
theta = np.ones(5)
```

```python

cost_matrix = []
alpha = 0.001
m = X.shape[0]
def sigmoid(z):
     return 1.0 / (1 + np.exp(-z))
for i in range(3000):
    z = (X*theta).sum(axis=1)
    h = sigmoid(z)
    cost_1 = -y*np.log(predict(X, theta))
    cost_2 = -(1-y)*np.log(1-predict(X, theta))
    cost = (1/X.shape[0])*(cost_1+cost_2).sum()
    print(cost, theta)
    cost_matrix.append(cost)
    for j in range(3):
        theta[j] = theta[j]-((alpha/m)*((h-y)*X.iloc[:, j]).sum())
```

```python

cost_matrix = []
lambdas = 1000
alpha = 0.001
m = X.shape[0]
def sigmoid(z):
     return 1.0 / (1 + np.exp(-z))
for i in range(2000):
    z = (X*theta).sum(axis=1)
    h = sigmoid(z)
    cost_1 = -y*np.log(predict(X, theta))
    cost_2 = -(1-y)*np.log(1-predict(X, theta))
    cost = (1/X.shape[0])*(cost_1+cost_2).sum() + ((lambdas/(2*X.shape[0]))*(theta**2).sum())
    print(cost, theta)
    cost_matrix.append(cost)
    for j in range(5):
        if j == 0:
            theta[j] = theta[j]-((alpha/m)*((h-y)*X.iloc[:, j]).sum())
        else:
            theta[j] = theta[j]-((alpha/m)*((h-y)*X.iloc[:, j]).sum()) - (((alpha*lambdas)/X.shape[0])*theta[j])
```

```python
predicted3 = h.apply(decision_boundary)
df['predicted3'] = predicted3
import seaborn as sns
sns.relplot(x='studied', y='slept',hue='predicted3' ,data=df )
```
