```python
def gradient_descent(X, y, theta, loop):
    cost_matrix = []
    m = X.shape[0]
    for i in range(loop):
        h = sigmoid(X, theta)
        cost_total = cost(X, y, theta, alpha)
        print(cost_total)
        cost_matrix.append(cost_total)
        for j in range(theta.shape[0]):
            theta[j] = theta[j]-((alpha/m)*((h-y)*X.iloc[:,j]).sum())
    return theta, cost_matrix
```
