```python
def minimumMoves(s, d):
    l = len(s)
    count = 0
    q = l-d+1
    st = 0
    e = d
    for i in range(q):
        print(i, st, e, s[st:e])
        
        if '1' not in s[st:e]:
            count += 1
        st = st+1
        e += 1
#         if e > l:
#             break
        print(count, '--')

    return count

minimumMoves('001000',3)
        
```
