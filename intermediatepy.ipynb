```python
def superout(x,y):
    print(x, y, 'super outt wrapper')
    def out(func):
        print('inside out')
        def inner(x):
            print('inside inner')
            return func(x)
        return inner
    return out


@superout(3,21)
def add(x):
    return x+2
# print(add.__name__)
add(232)
```

```python
def valid_func(func):
    def wrapper(x):
        if x > 0:
            if func(x) > 1:
                return 'value greater than 1'
            return 'value less than 1'
        else:
            return 'not valid function'
    return wrapper

@valid_func
def division_by(x):
    return 10/x
division_by(11)
```
