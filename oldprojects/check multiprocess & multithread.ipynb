```python
import concurrent.futures

import threading
import requests
import time

start = time.perf_counter()
def get_url(url):
    content = requests.get(url).content
    return content

url1 = 'https://openwho.org/courses/UNCT-COVID19-preparedness-and-response-EN'

with concurrent.futures.ThreadPoolExecutor() as executor:
    e1 = executor.submit(get_url, url1)
    print(e1)
#     print(e1.result())
print(e1)
stop = time.perf_counter()



print(stop-start)
```

```python
import time
import multiprocessing

start = time.perf_counter()
def func():
    for _ in range(100000000):
        pass
    print('inside function')

    
p1 = multiprocessing.Process(target=func)
p1.start()
p1.join()
p2 = multiprocessing.Process(target=func)
p2.start()
p2.join()


stop = time.perf_counter()

print(stop-start)
```
