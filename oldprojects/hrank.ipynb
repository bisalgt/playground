```python
# Sherlock and Squares


def find_square_numbers(a,b):
    count = 0
    for i in range(a, b+1):
        if i**(1/2)%2==0 or i**(1/2)%2==1:
            count += 1
    return count

find_square_numbers(24,49)
```

```python
# to improve timelimits

def find_square_numbers(a,b):
    count = 0
    sq_list = [i for i in range(a,b+1) if type(i**(1/2))==int]
    return len(sq_list)
find_square_numbers(24,49)
```

```python
import math
def find_square_numbers(a,b):
    count = 0
    for i in range(a, b+1):
        rem = math.sqrt(i)%2
        if rem == 0 or rem == 1:
            count += 1
    return count

%timeit find_square_numbers(178809602,330598161)
```

```python
def squares(a,b):
    for i in range(a, b+1):
        sq_root = i**(1/2)
        rem = sq_root%2
        if rem ==0 or rem ==1:
            break
    count = 0
    print(sq_root)
    for j in range(int(sq_root),b+1):
        if j**2  in range(a, b+1):
            count += 1
        if j**2 not in range(a, b+1):
            break
    return count

squares(178809602,330598161)
```

```python
def is_leap_year(yr):
    if yr%4 == 0:
        if yr%100 != 0:
            if yr%400 == 0:
                return True
        return True
    return False

is_leap_year(2000)
```
