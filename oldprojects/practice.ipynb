```python
import unittest

class CheckStringMethods(unittest.TestCase):
    
    @classmethod
    def setUpClass(cls):
        print('inside setUpClass')
        
    @classmethod
    def tearDownClass(cls):
        print('inside tearDownClass')
    
    def setUp(self):
        print('Iniside test fixture: setup')
        self.title = 'this is the title'
    
    def test_lower(self):
        print('Iniside test case: test_lower')
        self.assertEqual(self.title.lower(), 'this is the title', msg='testing if the test is lower')
    
    def test_upper(self):
        print('Inside test case: test upper')
        self.assertEqual(self.title.upper(), 'THIS IS THE TItle', msg='testing if the test is upper')
    
    def tearDown(self):
        print('Inside test fixutre: tearDown')
        del self.title

# if __name__ == '__main__':
#     CheckStringMethods()
```

```python
suite = unittest.TestSuite()
suite.addTests([CheckStringMethods('test_lower'), CheckStringMethods('test_upper')])

runner = unittest.TextTestRunner()
runner.run(suite)
```
