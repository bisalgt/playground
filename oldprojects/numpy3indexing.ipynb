# INDEXING IS ONLY POSSIBLE BY INTEGERS NOT BY FLOATING VALUES

```python
i = np.array([1,2,3,1,])
```

```python
x[i] # passing array as index
```

```python
x[[1,2,3,1]]
```

### NumPy arrays may be indexed with other arrays (or any other sequence- like object that can be converted to an array, such as lists, with the exception of tuples;

```python
x[(1,2,3,1)]
```

### . For all cases of index arrays, what is returned is a copy of the original data, not a view as one gets for slices.

#### Negative values are permitted and work as they do with single indices or slices:

```python
x[np.array((1,-1,1,1))]
```

```python
x[[[1,1],[3,1]]]
```

```python
x
```

```python
x[np.array([np.arange(4).reshape(2,2)])]
```

```python
np.arange(4).reshape(2,-1)
```

```python
x[np.arange(4).reshape(2,-1)]
```

```python
x[np.arange(3), np.arange(3)]
```

```python
x[[np.arange(3), np.arange(3)], [np.arange(3), np.arange(3)]]
```

```python
x[np.array([0,2,4]), np.array([0,1])]
```

```python
x[np.array((0,1,3))]
```

# What results is the construction of a new array where each value of the index array selects one row from the array being indexed and the resultant array has the resulting shape (number of index elements, size of row).

## Boolean arrays must be of the same shape as the initial dimensions of the array being indexed

```python
b = x > 20
```

```python
b
```

##### boolean indexing

```python
x[b]
```

## Unlike in the case of integer index arrays, in the boolean case, the result is a 1-D array containing all the elements in the indexed array corresponding to all the true elements in the boolean array.

```python
x[np.nonzero(b)]
```

### both x[b] as x[np.nonzero(b)] is same

```python
b[:,5]
```

```python
x[b[:,5]]
```
