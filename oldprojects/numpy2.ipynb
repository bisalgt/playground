### Skipping lines and choosing columns
The skip_header and skip_footer arguments

```python
data = u"""
# this is comment
id, age, salary
1,21,23433
2,32,44444
3,44,488888
4,2,44
5,6,77

"""
```

```python

np.genfromtxt(StringIO(data), delimiter=',')
```

```python
np.genfromtxt(StringIO(data), delimiter=',', skip_header=1)
```

```python
np.genfromtxt(StringIO(data), delimiter=',', skip_header=3)
```

```python
np.genfromtxt(StringIO(data), skip_footer=1, skip_header=3, delimiter=',', comments='#', autostrip=True)
```

### The usecols argument

```python
data = u"1,2,3,4,5\n2,3,4,5,6\n4,4,4,4,4"
```

```python

np.genfromtxt(StringIO(data), delimiter=',')
```

```python
np.genfromtxt(StringIO(data), delimiter=',', usecols=(0,1, -1, -2, 0))
```

## Choosing the data type¶

```python
data = u"1, 2, 3 \n 4 ,5 ,6 "
```

```python

np.genfromtxt(StringIO(data), names="'a','b','c'", delimiter=',')
```

```python
np.genfromtxt(StringIO(data), dtype=('a', int))
```

```python
np.genfromtxt(data, dtype=[(_, int) for _ in "abc"], delimiter=None)
```

```python
data = StringIO("1 2 3\n 4 5 6")
```

```python
np.genfromtxt(data, dtype=[(_, int) for _ in "abc"])
```

```python
data = StringIO("1 mis 3 ?\n 4 5 6 ?\n 7 mis N/A 10")
np.genfromtxt(data, missing_values={1:'mis', 2:'N/A', 3:'?'}, filling_values={1:1, 2:2, 3:3})
```
