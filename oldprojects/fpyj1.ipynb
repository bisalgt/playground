```python
class Person:
    def __init__(self, first, last):
        self.first = first
        self.last = last
    
    @property
    def email(self):
        return f'{self.first}.{self.last}@email.com'
    
    @email.setter
    def email(self, mail):
        self.first = mail.split('.')[0]
        self.last = mail.split('.')[-1]
        
```

```python
class Person:
    def __init__(self, first, last):
        self.first = first
        self.last = last
    
    @property
    def email(self):
        return f'{self.first}.{self.last}@email.com'
    
    @email.setter
    def email(self, mail):
        self.first = mail.split('.')[0]
        self.last = mail.split('.')[-1]
    
    @email.deleter
    def email(self):
        self.first = None
        self.last = None
        
```
