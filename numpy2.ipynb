# Working with IO

## Importing data with genfromtxt

```python
import numpy as np
from io import StringIO
```

```python
data = u"1, 2, 3, 4\n5, 6, 7, 8"
```

```python
type(data)
```

```python
repr(data)
```

```python
np.genfromtxt(StringIO(data), delimiter=',')
```

###### The delimiter keyword is used to define how the splitting should take place

```python
np.genfromtxt(StringIO('datas/csvtorandomize.csv'), delimiter=',')
```
