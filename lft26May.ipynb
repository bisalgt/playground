```python
def logic(my_input):
    s1 = my_input.split(' ')
    s = sorted(s1, key=len, reverse=True)
    m = len(s[0])
    ch = my_input[0]
    for indx, val in enumerate(s1):
        if len(val) < m:
            temp = val.rjust(m, ch)
            s1[indx] = temp
    r = ' '.join(s1)
    return r
logic('p am hellooooo bishal gau')
```
