#### bisect as a faster replacement for the index method when searching
#### through long ordered sequences of numbers

```python
s = list('edcba')
from bisect import insort
insort(s,'d')
print(s)
s = list('abcde')
insort(s, 'd')
print(s)
```

```python
from bisect import insort

raw = list('mynameisbishalgautam')
li = []
for i in raw:
    insort(li, i)
print(''.join(li))
```
