```python
def factorial(n):
    if n == 1:
        return 1
    mul = 1
    for i in range(1,n+1):
        mul *= i
    return mul

factorial(5)
    
```

```python
def div_5(x):
    s = str(x)
    d = 0
    for i,j in enumerate(s[::-1]):
        d += int(j)*(2**i)
    if d%5 == 0:
        return True
    return False
div_5(1111)
```

```python
def rotate(li, k):
    if k%len(li) == 0:
        return li
    k = k%len(li)
    r = li[k:]+li[:k]
    return r
rotate('123456',36)
```

```python
def check_brackets(s):
    p = ['(', '{', '[']
    n = [')', '}', ']']
    count = 0
    for i in s:
        if i in p:
            count += 1
        elif i in n:
            count -= 1
    if count ==0:
        return 'Valid'
    return 'Invalid'
check_brackets('{{[]}}')
```
