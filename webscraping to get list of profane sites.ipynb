```python
list_link = []
with open('profane_links.txt', 'a') as wf:
    for tag in soup.find_all('li'):
        link = tag.a['href']
        print(link)
        if link.__contains__('.com'):
            indx = link.index('.com')
            link = link[:indx+4]
        if link.__contains__('.net/'):
            indx = link.index('.net/')
            link = link[:indx+4]
        if link.__contains__('.live/'):
            indx = link.index('.live/')
            link = link[:indx+5]
        if link.__contains__('.link/'):
            indx = link.index('.link/')
            link = link[:indx+5]
        if link.__contains__('.re/'):
            indx = link.index('.re/')
            link = link[:indx+3]
        if link.__contains__('.org'):
            indx = link.index('.org')
            link = link[:indx+4]
        link.strip()
        list_link.append(link)
        wf.write(link)
        wf.write("\n")
       
    
```
